```python
#basic datastructures : integer,float,boolean,string (not derived data structures like lists,tuple,dictionaries,sets are mutable) are IMMUTABLE
```

```python
# Original string
my_string = "hello"
print(f"Original string: {my_string}, Memory address: {id(my_string)}")

# Attempting to "modify" the string
new_string = my_string.replace('h', 'j')
print(f"New string after replace: {new_string}, Memory address: {id(new_string)}")

# Concatenation also creates a new string
another_string = my_string + " world"
print(f"Concatenated string: {another_string}, Memory address: {id(another_string)}")
```
