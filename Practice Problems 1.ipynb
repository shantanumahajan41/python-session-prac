```python
def checkprime():
    
    try :
        numb = int(input("Enter a number : "))
        if numb <= 0:
            print("Enter a Positive number !!")
            checkprime()
        elif numb == 2:
            print("2 is prime.")
        elif numb == 1 :
            print("1 is neither prime nor composite")
        elif numb != 2 and numb >= 1 :
            factors = [1, numb]
            for i in range(2, numb):
                if numb % i == 0:
                    factors.append(i)
                else :
                    pass
            if len(factors) > 2:
                print(f"The number is not prime as its factors are : {factors}") 
            else:
                print("The number is Prime!")
        print("Hope you enjoyed the program")
    except ValueError:
        print("Please Enter a Valid number !!")
        checkprime()

    
       
        
           
```
