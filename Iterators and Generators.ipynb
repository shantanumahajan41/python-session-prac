# Generators

```python
def square_numbers(nums):
    result = []
    for i in nums:
        result.append(i*i)
    return result
```

```python
my_nums = square_numbers([1,2,3,4,5])
```

```python
print(my_nums)
```

```python
def square_numbers(nums):
    for i in nums:
        yield (i*i)
```

```python
my_nums = square_numbers([1,2,3,4,5])
```

```python
print(my_nums)
```

```python
print(next(my_nums))
print(next(my_nums))
print(next(my_nums))
print(next(my_nums))
print(next(my_nums))
```

```python
#list comprehension
my_nums = [x*x for x in [1,2,3,4,5]]
print(my_nums)
for num in my_nums:
    print(num)
```

```python
# creating its generator
my_nums = (x*x for x in [1,2,3,4,5])
print(my_nums)
for num in my_nums:
    print(num)
```

```python
my_nums = (x*x for x in [1,2,3,4,5])
print(my_nums)
print(list(my_nums))
```
