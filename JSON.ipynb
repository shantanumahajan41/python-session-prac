```python
import json

people_string = '''
{
  "people": [
    {
      "name": "John Smith",
      "phone": "615-555-7164",
      "emails": ["johnsmith@bogusemail.com", "john.smith@work-place.com"],
      "has_license": false
    },
    {
      "name": "Jane Doe",
      "phone": "560-555-5153",
      "emails": null,
      "has_license": true
    }
  ]
}
'''


data = json.loads(people_string)

for person in data['people']:
    del person['phone']

new_string = json.dumps(data, indent = 2)

print(new_string)
```

```python
import json

people_string = '''
{
  "people": [
    {
      "name": "John Smith",
      "phone": "615-555-7164",
      "emails": ["johnsmith@bogusemail.com", "john.smith@work-place.com"],
      "has_license": false
    },
    {
      "name": "Jane Doe",
      "phone": "560-555-5153",
      "emails": null,
      "has_license": true
    }
  ]
}
'''


data = json.loads(people_string)

for person in data['people']:
    del person['phone']

new_string = json.dumps(data, indent = 2, sort_keys = True)

print(new_string)
```

```python
#'load' method in json loads a file while 'loads' method loads a string

import json 

with open('myjson.json') as f:
    data = json.load(f)

    for states in data['us_states']:
        print(states)
```

```python
import json 

with open('myjson.json') as f:
    data = json.load(f)

for states in data['us_states']:
    print(states['state_name'], states['abbreviation'])
```

```python
import json 

with open('myjson.json') as f:
    data = json.load(f)

for states in data['us_states']:
    del states['area_codes']

#similar to difference between 'load' and 'loads' ,here for 'dump' and 'dumps'

with open('new_states.json', 'w') as f:
    json.dump(data, f)
```
