```python
#lets make list of squares of numbers 
a = []
for i in range(10):
    a.append(i**2)
```

```python
print(a)
```

#### List Comprehension

```python
[i**2 for i in range(10)]
```

```python
#this can be further made...like if we want list of sqaures of only even numbers from 0-9
```

```python
[i**2 for i in range(10) if i % 2 == 0]
```

#### Dict Comprehension

```python
#lets make a dictionary where key is number and its value is its square
```

```python
{ i: i**2 for i in range(10) }
```

```python
#now similar to previous
{ i: i**2 for i in range(10) if i % 2 ==0 }
```

#### Set Comprehension

```python
{ i**2 for i in range(10)}     #not key "i:" this time 
```

 #### Tuple Comprehension ?

```python
a = (i**2 for i in range(10))
```
