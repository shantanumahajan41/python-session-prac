```python
import csv

with open('mycsvfile.csv' ,'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    with open('new_names_dict.csv', 'w') as newfile:
        fieldnames = ['first name', 'last name', 'email']
        csv_writer = csv.DictWriter(newfile, fieldnames = fieldnames, delimiter = '\t')

        csv_writer.writeheader()
        
        for row in csv_reader:
            csv_writer.writerow(row)
```

```python
import csv

with open('mycsvfile.csv' ,'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    with open('new_names_dict.csv', 'w') as newfile:
        fieldnames = ['first name', 'last name', 'email']
        csv_writer = csv.DictWriter(newfile, fieldnames = fieldnames, delimiter = '\t')

        csv_writer.writeheader()
        
        for row in csv_reader:
            del row['email']
            csv_writer.writerow(row)         #title 'email' will be there but all emails are deleted
```
