# Dictionaries
- A dictionary is like an address book where you can find the address or contact details of a person by knowing only his/her name i.e. we associate keys (name) with values (details). Note that the key must be unique just like you cannot find out the correct information if you have two persons with the exact same name.
- Remember that the key - value pair in dictionary are not ordered in any manner.

```python
#in lists and tuples, information is accessed using indexes, hence they are called as ordered databases.
```

```python
a = [1, 2, 3, 4]
```

```python
a[0]
```

```python
#dictionaries are unordered
```

```python
a = {
    "name": "jatin",
    "marks": 90,
    "subjects": ["eng", "math"],
    "friends": {
        "pulkit": "android",
        "gurpeet": "iphone"
    }
}
```

```python
a["name"]
```

```python
a["subjects"]
```

```python
a["friends"]
```

```python
for key in a:
    print(key)
```

### Some important dict methods :-
- get()
- items()
- keys()
- values()
- clear()

```python
 for key in a.items():
     print(key)
```

```python
for key, value in a.items():
    print(key, "=>", value)
```

```python
for key in a.keys():
    print(key)
```

```python
for val in a.values():
    print(val)
```
