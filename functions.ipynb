```python
#Now this function takes another function as argument and wraps it in such a way that it will require a authentication...
```

```python
def login_required(func):
    def wrapper(username, password, *args, **kwargs):
        if username in users and users[username] == password:
            # User is authenticated
            func(*args, **kwargs) #tuple gets inflated due to *argument
        else:
            print("Not Authenticated")
    return wrapper        
```

```python
def add(a, b):
    print(a+b)
```

```python
add(1, 2)
```

```python
protected_add = login_required(add)
```

```python
print(protected_add)
```

```python
protected_add()
```

```python
protected_add("jatin", "password")
```

```python
protected_add("jatin", "password", 1, 3)
```

```python
protected_add("jatin", "password123", 2, 3)
```
